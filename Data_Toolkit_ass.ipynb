THEORY QUESTION

1.What is NumPy, and why is it widely used in Python?

ANS.1. NumPy (Numerical Python) is a library for fast numerical computing. It provides multidimensional arrays and functions for mathematical operations.

It is Widely used because it’s fast, memory-efficient, and has a rich set of linear algebra, statistical and random number tools.

2.How does broadcasting work in NumPy?

ANS.2. Broadcasting lets NumPy apply operations between arrays of different shapes without explicit looping.
Example:
a = np.array([1,2,3])
b = 2
print(a + b)

3.What is a Pandas DataFrame?

ANS.3. A DataFrame is a 2D labeled data structure (rows & columns), similar to an Excel table or SQL table, built on top of NumPy arrays.

4. Explain the use of the groupby() method in Pandas?

ANS.4. groupby() splits data into groups based on column values, applies a function (like mean, sum), and combines results.
df.groupby('Deptt')['Marks'].mean()

5. Why is Seaborn preferred for statistical visualizations?

ANS.5. Seaborn is built on Matplotlib and automatically handles themes, color palettes, and statistical plotting (like boxplots, violin plots). It’s easier and prettier for data exploration.

6. What are the differences between NumPy arrays and Python lists?

ANS.6.
Feature      NumPy Array              Python List   
Speed        Faster (C-based)         Slower        
Operations   Vectorized               Loops needed  
Memory       Fixed, less overhead     More overhead
Type         Homogeneous              Heterogeneous

7. What is a heatmap, and when should it be used?

ANS.7. A heatmap is a graphical representation of data values with colors. Used to show correlations, matrix data, or intensity patterns.

8. What does the term “vectorized operation” mean in NumPy?

Applying an operation on entire arrays at once (element-wise) without explicit Python loops. This uses underlying C routines → much faster.

9. How does Matplotlib differ from Plotly?

ANS.9. Matplotlib: static, publication-quality plots, very flexible, no built-in interactivity.

Plotly: interactive, web-friendly, zoom/pan/hover by default.

10. What is the significance of hierarchical indexing in Pandas?

ANS.10. Hierarchical (multi-level) indexing lets you work with higher-dimensional data in 2D tables, e.g. rows indexed by (region, year). Useful for complex grouping and pivotin

11. What is the role of Seaborn’s pairplot() function?

ANS.11. It plots pairwise relationships across an entire DataFrame. Great for seeing scatterplots and distributions for each pair of variables quickly.

12. What is the purpose of the describe() function in PandaS?

ANS.12. It provides summary statistics (count, mean, std, min, max, quartiles) for numeric columns in one go.

13. Why is handling missing data important in Pandas?

ANS.13. Missing data can skew analysis and models. Pandas has isna(), fillna(), dropna() to detect and handle missing values to ensure reliable results.

14. What are the benefits of using Plotly for data visualization?

ANS.14. Interactive dashboards, 3D plots, map visualizations, works well with web apps (Dash). Users can zoom and export easily.

15. How does NumPy handle multidimensional arrays?

ANS.15.It uses the ndarray object which can be n-dimensional. Each dimension has a shape and stride. Operations are generalized to N-D arrays (matrix, tensor, etc.).

16. What is the role of Bokeh in data visualization?

ANS.16. Bokeh produces interactive, browser-ready visualizations with Python. Great for dashboards and streaming data.

17. Explain the difference between apply() and map() in Pandas?

ANS.17.Series.map(func): element-wise on a Series.

DataFrame.apply(func): apply along axis (rows or columns) of a DataFrame.

18.  What are some advanced features of NumPy?

ANS.18. Broadcasting, masked arrays, structured arrays (records), memory-mapping, linear algebra (linalg), random sampling.

19. How does Pandas simplify time series analysis?
ANS.19. It has built-in DateTime index, resampling, shifting/lagging, rolling windows, time zone handling, making time-based grouping very easy.

20.  What is the role of a pivot table in Pandas?

ANS.20.A pivot table in Pandas is used to summarize, aggregate, and restructure data—similar to Excel’s Pivot Table.
It allows you to group data by one or more keys (rows/columns) and apply an aggregation function (like mean, sum, count, etc.) to get a condensed view of the dataset.

21.  Why is NumPy’s array slicing faster than Python’s list slicing?

ANS.21 NumPy arrays are stored in contiguous blocks of memory and slicing creates views (no data copy), whereas list slicing makes a new list and copies elements.
